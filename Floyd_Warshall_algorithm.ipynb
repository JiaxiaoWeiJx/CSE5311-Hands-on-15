# **Floyd-Warshall algorithm**

- Nodes: 1, 2, 3, 4, 5
- Edges and weights:
 - (1, 2) with weight 3
 - (1, 5) with weight -4
 - (1, 3) with weight 8
 - (2, 5) with weight 7
 - (2, 4) with weight 1
 - (3, 2) with weight 4
 - (4, 3) with weight -5
 - (4, 1) with weight 2
 - (5, 4) with weight 6

```python
def floyd_warshall(graph):
    # Find the maximum node label
    max_node = max(graph.keys())

    # Initialize the distances matrix with the same weights as the graph
    # If there's no edge between two nodes, set the distance to infinity
    distances = [[float('inf')] * max_node for _ in range(max_node)]
    for i in range(max_node):
        distances[i][i] = 0
    for i in graph:
        for j in graph[i]:
            distances[i-1][j-1] = graph[i][j]

    # Floyd-Warshall algorithm
    for k in range(max_node):
        for i in range(max_node):
            for j in range(max_node):
                if distances[i][k] + distances[k][j] < distances[i][j]:
                    distances[i][j] = distances[i][k] + distances[k][j]

    return distances

# Define the graph
graph = {
    1: {2: 3, 5: -4, 3: 8},
    2: {5: 7, 4: 1},
    3: {2: 4},
    4: {3: -5, 1: 2},
    5: {4: 6}
}

# Find the shortest distances between all pairs of nodes
shortest_distances = floyd_warshall(graph)

# Print the shortest distances in a 5 by 5 matrix format
print("Shortest distances matrix:")
for row in shortest_distances:
    print(" ".join(str(dist) if dist != float('inf') else 'âˆž' for dist in row))

```
